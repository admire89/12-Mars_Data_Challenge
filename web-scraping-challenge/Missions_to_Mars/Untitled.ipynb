```python
#Dependecies 
from bs4 import BeautifulSoup 
import pymongo 
import time 
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
executable_path
```

```python
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.verywellmind.com/istj-introversion-sensing-thinking-judgment-2795992'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
image_div =soup.find_all("div", class_ ="figure-media")[0]
image_link = image_div.img['data-src']
```
